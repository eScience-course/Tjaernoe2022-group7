```python
import xarray as xr
xr.set_options(display_style='html')
import intake
import cftime
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
%matplotlib inline
```

Catalog with all CMIP6 experiments stored in the google server

```python
cat_url = "https://storage.googleapis.com/cmip6/pangeo-cmip6.json"
col = intake.open_esm_datastore(cat_url)
col
```

Find all possible variables, experiments etc...

```python
# list all the variable_id in col
print(col.df.variable_id.unique())
# you change variable_id to other parameters
```

Search in the catalog

```python
# example: CESM2 model, historical experiment, tasmax variable
cat = col.search(source_id=['CESM2'], experiment_id=['historical'], table_id=['Amon'], variable_id=['tas'], member_id=['r1i1p1f1'])
```

Create dictionary from the list of datasets we found

```python
dset_dict = cat.to_dataset_dict(zarr_kwargs={'use_cftime':True})
```

Open dataset

```python
dset = dset_dict['CMIP.NCAR.CESM2.historical.Amon.gn']
dset
```

Convert time from cftime.DatetimeNoLeap to datetime (easier to work with)

```python
dset['time'] = dset['time'].to_dataframe().index.to_datetimeindex()
```
